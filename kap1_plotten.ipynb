## Der Kernel

Ein Notebook ist etwas anderes als ein Programm oder ein Skript. Jede Zelle des Codes wird bei der Ausführung, d.h. bei Shift-Enter, an den Kernel gesendet und dort ausgewertet. Der Kernel merkt sich alle Variablen und gibt das Ergebnis der letzten Zeile an das Notebook zurück, das es anzeigt.

Die Reihenfolge der Zellen hängt davon ab, wie sie ausgeführt werden. Es ist auch möglich, vorherige Zellen erneut auszuführen. In diesem Fall kann sich das Ergebnis aber ändern, da sich der Kernel alle vorherigen Ausführungen merkt.

Hier ist ein Beispiel: Wenn Sie die folgende Zelle ausführen, erhalten Sie zunächst das Ergebnis '10' aus dem obigen Beispiel. Der Kernel erinnert sich noch daran.

````python
x
````

Wenn Sie dann die nächste Zelle ausführen, erhalten Sie '11'. Wenn Sie dann die vorherige Zelle wieder ausführen, erhalten Sie 11 und nicht 10.

````python
x = x + 1
x
````

Die Reihenfolge, in der die Zellen ausgeführt werden, bestimmt also das Ergebnis, ist aber im Notebook selbst nicht dokumentiert. Dies kann zu unerwünschten Effekten führen. Es ist daher sinnvoll, im Zweifelsfall den Kernel neu zu starten und alles in der Reihenfolge des Notebooks auswerten zu lassen (via Kernel - restart and run...). Das gibt es darum auch als Doppelpfeil-Icon oben am Fenster.

## Dateien laden

Messdaten werden üblicherweise (und so auch von Pasco Capstone) als reine Textdatei (ASCII) im 'CSV'-Format gespeichert. CSV steht für 'coma separated values', die einzelnen Spalten sind also durch Kommata getrennt und eine Zeile enthält einen zusammenhängenden Datenpunkt, z.B. alle Messkanäle zu einem Zeitpunkt.

Das Komma als Spaltentrenner hat sich aber nicht wirklich durchgesetzt. Man findet auch Semikolons, Leerzeichen oder Tabulatoren (\t). Pasco verwendet ein Semikolon. Es ist auch darauf zu achten, was als Dezimaltrennzeichen verwendet wird. Auf Systemen mit deutscher Benutzeroberfläche ist es meist ein Komma, sonst eher ein Punkt. 

All diese Informationen müssen beim Laden einer Datei angegeben werden. Relativ einfach geht das mit `csv_read` aus dem Paket `pandas`.

## Graphen zeichnen = plotten

Zum Zeichnen von Graphen eignet sich die Bibliothek 'matplotlib'.


````python
import matplotlib.pyplot as plt # zum plotten
````

Es sind eigentlich immer die selben Befehle, die man dann braucht, einen Graphen zu zeichnen. Hier ein Beispiel mit Erklärung in den Kommentaren.

````python
plt.plot(data.t, data.alpha, color="red", label="Messung")  # plottet den Winkel eines Pendels
# plt.plot(...)                                             # weitere Linie in das selbe Koordinatensystem 
plt.xlabel("t (s)")                                         # Achsenbeschriftung
plt.ylabel("alpha (rad)") 
plt.title("Pendel")                                         # Titel
plt.grid()                                                  # Gitterlinien
plt.legend(loc='upper left' )                               # zeigt die Legende, `legend()` positioniert automatisch
# plt.savefig("pendel.png")                                 # speichert das Bild als png oder pdf
````

Man kann den Ausschnitt anpassen

````python
plt.plot(data.t, data.alpha, color="red", label="Messung") # plottet den Winkel eines Pendels
plt.xlim(3, 5) # x-Achsenbereich
plt.ylim(-1, 10) # y-Achsenbereich
````

oder die Linieart, Dicke, Farbe oder Symbole

````python
plt.plot(data.t, data.alpha, color="blue", linestyle="--", marker="o", markersize = 5) # plottet den Winkel eines Pendels
````

Es ist auch möglich, mehrere Graphen in ein Bild zu zeichnen. Das wird dann aber bald aufwändig.

````python
# Create a figure with two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(7, 4))

# First plot
axes[0].plot(data.t, data.alpha, color="blue", label="Messung 1")
axes[0].set_title("Plot 1")
axes[0].set_xlabel("t (s)")
axes[0].set_ylabel("alpha (rad)")

# Second plot
axes[1].plot(data.t, data.alpha, marker='o', markersize=3, linestyle='none',label="Messung 2")
axes[1].set_title("Plot 2")
axes[1].set_xlabel("t (s)")
axes[1].set_ylabel("alpha (rad)")

# Adjust layout
plt.tight_layout()   # entfernt den Überlapp
````

Hilfslinien und Beschriftungen kann man auch hinzufügen.

````python
plt.plot(data.t, data.alpha, color="blue") # plottet den Winkel eines Pendels
plt.axhline(0.6) # horizontale Linie bei y=0.6
plt.axvline(2) # vertikale Linie bei x=2
plt.axline( (2, 0.6), slope=-1, color="green", label="drop") #  Linie bei x=2 und y=0.6 mit Steigung -1
plt.text(2, 0.6, "Hier ist die Linie", fontsize=12) # Text an der Stelle (2,0.6)
````

## Gute Graphen

Das Zeichnen von Graphen wird Sie von nun an, zumindest während Ihres Studiums, wahrscheinlich aber Ihr ganzes Leben lang begleiten. Nicht nur in der Experimentalphysik, sondern auch in der computergestützten theoretischen Physik geht es letztlich darum, eine Aussage durch eine graphische Darstellung von Zahlen zu untermauern. Es lohnt sich also, darüber nachzudenken, was einen guten Graphen ausmacht. Nicht jedes Ergebnis von `plt.plot()` ist gut!

Es gibt grundlegende Anforderungen an einen Graphen:
* passende gewählte Achsen-Skalierung und Intervalle der Skala
* beschriftete Achsen im Format "Zeit t (s)", "Winkel alpha (Grad)" oder  [ ] statt ( )
* beschriftete Kurven, Symbole und Hilfslinien

Es hilft zuerst zu überlegen, was denn die **Aussage** eines Graphen sein soll. Fassen Sie in einem Satz zusammen, was Sie mit dem Graphen sagen wollen, beispielsweise
* "Die Reibung des Pendels ist vernachlässigbar"
* "Die Periode des Pendels ist konstant"


Optimieren Sie dann die Darstellung auf diese Aussage hin und lassen Sie alles andere weg, was von der Aussage ablenken könnte. Für das Gitter (grid) gibt es wirklich selten einen guten Grund.

In diesem Sinne sind alle obigen Darstellungen der Pendel-Daten **keine guten Graphen**. Der erste (mit der Achsenbeschriftung) würde zu der Aussage passen "wir haben ein Pendel gemessen".

Das Standardwerk zur Darstellung von Informationen in Graphen ist *Edward R. Tufte: The visual display of quantitative information*, ausleihbar in der Bibliothek.

Hier ein Beispiel für einen Graphen zur Aussage "Die Reibung des Pendels ist vernachlässigbar"

````python
plt.plot( (data.t -2) / 0.9, data.alpha / 0.598, color="red") # plottet den Winkel eines Pendels
plt.xlabel("Nummer des Umlaufs") # x-Achsenbeschriftung
plt.ylabel("normierte Auslenkung") # y-Achsenbeschriftung
plt.title("Quasi reibungsfreies Pendel") # Titel

plt.axhline(1)
plt.axline( (0, 1), slope=-0.005, color="green", linestyle="--") #  
plt.text(5, 0.98, "0.5% Verlust pro Umlauf", fontsize=10) 
plt.ylim(0.9, 1.03)
plt.xlim(-0.4, 9.4)
````
